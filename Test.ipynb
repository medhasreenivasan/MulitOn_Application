```python
import gradio as gr
import requests

def upload_image_and_get_result(image, user_input):

    files = {"image": open(image,"rb")}
    data = {"user_input": user_input}

    # Send the POST request to the endpoint
    response = requests.post(url, files=files, data=data)

    # Check if the request was successful
    if response.status_code == 200:
        # Return the result
        return response.json()["browse"]
    else:
        return "Error: Unable to process the request."

# Define the Gradio interface
inputs = [
    gr.Image(label="Upload Image",type='filepath'),
    gr.Textbox(label="command")  # User input text box
]
outputs = gr.Textbox(label="Result")

# Create the Gradio interface
gr.Interface(upload_image_and_get_result, inputs, outputs, title="Image Upload and Processing").launch(debug=True)
```
